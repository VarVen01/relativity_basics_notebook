# Introduction to Relativity and the Lorentz Factor
<p align = "justify"> Everything depends on your frame of reference. For example, say you are in a fast-moving train and your friend is standing outside on the sidewalk to see you off. In your friend's point of view, you are moving past them while they are standing still. However, in your point of view, your friend is moving past you while you are standing still!

<p align = "justify"> Relativity is the study of how observers in different frames of reference moving with respect to each other measure the same event. Though Galileo and Newton laid the foundations of classical physics, Einstein was the one who developed the modern theory of relativity in the 1900s.

<p align = "justify"> This modern relativity is divided into two parts: <i>Special Relativity</i> (1905), which deals with frames of reference moving at constant velocity, and <i>General Relativity</i> (1916), which deals with frames of reference moving at constant acceleration.

What's more, Einstein's theories aren't just predictions; they have in fact been verified experimentally as well.

The Lorentz factor (symbolized by gamma) is used to convert between non-relativistic and relativistic measurements. Its formula is given as

<p align = "center"> $γ = \frac{1}{\sqrt{ 1 - \frac{v^2}{c^2}}}$.

In the below code snippet, let's import all the libraries we're going to need first. We can also set c equal to the speed of light (m/s) because it will be useful later on.

````python
import numpy as np
import matplotlib.pyplot as plt

c = 3e8
````

# Postulates of Special Relativity

1. The laws of physics are the same in all inertial reference frames.
2. The speed of light is constant in all inertial reference frames and is unaffected by the speed of its source.

An inertial reference frame is one where all objects follow Newton's first law of motion (objects at rest remain at rest, and objects in motion remain in motion at constant velocity in a straight line, unless acted upon by an external force).

The speed of light in a vacuum is approximately $3*10^8$ m/s.

# Special Relativity: Time Dilation
Time dilation is a phenomenon in which moving clocks (traveling at very high speeds) appear to be running slower than stationary clocks. Here's a fun story: how did Einstein begin coming up with the theory of relativity in the first place? It's said that one day he was traveling by train and looked out the window to see a giant clock tower. He wondered, 'what would the time look like if this train was traveling extremely fast?' And the rest is history!

The relationship between the non-relativistic time and the relativistic time is given as follows:

<center> $t = γ*t_0$,

<p align="justify"> where $t_0$ is the time measured in the moving frame. Notice that the factor gamma is always greater than 1 (because v is always less than c). Therefore, each inertial observer sees that all clocks in motion relative to that observer run slower than the observer's clock.

<p align="justify"> For example, say Person A is on a train traveling at relativistic speed and Person B is standing on the train platform. Person B is holding a clock, and the train has clocks on each traincar. Person B will view the clocks on the train as running slower than their clock. But, in Person A's reference frame, Person A is at rest and the train platform is the one traveling at relativistic speeds. So Person A will view Person B's clock as running slower than the clocks on the train!

<p align="justify"> Let's code an example and plot it for better visualization. In the below code snippet, we define the proper time and use it to find the dilated time using the formula for time dilation given above. By specifying a range for the values of velocity v (1 to 0.99c), we can plot dilated time as a function of v. You can also try this out for different proper times by changing the value of t_proper!

````python
def time_dilation(v, t_proper, c):
  gamma = 1/np.sqrt(1-(v**2/c**2))
  return gamma*t_proper

v_values = np.linspace(1, 0.99*c, 100)
t_proper = 1
t_dilated = time_dilation(v_values, t_proper, c)

plt.plot(v_values /c, t_dilated, label = "Time Dilation")
plt.xlabel('Velocity as a Fraction of Speed of Light')
plt.ylabel('Dilated Time (s)')
plt.legend()
plt.show()
````

# Add-On: Twin Paradox
<p align="justify"> Let's say there are two twins on Earth. Of course, on Earth they are the same age. Twin A remains on Earth, while Twin B boards a spaceship and begins traveling the galaxy with a speed close to the speed of light. For Twin B, only a few months may have passed on their journey, but depending on how fast the spaceship is going, it may have been decades for Twin A.

But this isn't really a paradox: this can be explained by the phenomenon of time dilation that we visited earlier.

<p align="justify"> The interesting part is really due to frames of reference. In Twin A's reference frame, Twin B is the one moving away at relativistic speed. But in Twin B's reference frame, <i>Twin A</i> is the one moving away at relativistic speed. So who's correct, Twin A or Twin B?

<p align="justify"> How do we solve this paradox? We need to look at the overall picture. As long as Twin B is traveling at constant speed, the twins will fundamentally disagree about the flow of time, and both perspectives will be valid. However, Twin B has to leave Earth, as well as come back. So, at certain points, Twin B needs to slow down, stop, turn around, and accelerate. Due to this asymmetry, Twin B can no longer claim to be in an inertial frame of reference because they will know they were moving (ex. from vibrations on the ship as the velocity changes). When Twin B returns to Earth, they know that they were the traveling twin.

<p align = "justify"> In the end, we see that one can calculate the ages of the twins using the formula for time dilation, and the twins will also agree on these ages when they eventually compare notes. Let's code an example and plot the ages of Twin A on Earth and Twin B traveling at a relativistic speed. The proper time is the time experienced by Twin B while on the spaceship. As  Twin B travels at faster speeds, Twin A perceives many more years passing than Twin B; for instance, while Twin B may only be a young adult upon returning, Twin A may be an elderly person!

````python
def time_dilation(v, t_proper, c):
  gamma = 1/np.sqrt(1-(v**2/c**2))
  return gamma*t_proper

v_values = np.linspace(1, 0.99*c, 100)
t_proper = 25

t_dilated = time_dilation(v_values, t_proper, 3e8)

plt.plot(v_values / 3e8, t_dilated, label="Twin A (Stationary)")
plt.axhline(t_proper, color='r', linestyle='--', label="Twin B (Moving)")
plt.xlabel("Velocity as a Fraction of c")
plt.ylabel("Years Passed")
plt.legend()
plt.show()
````

# Special Relativity: Length Contraction
Length contraction refers to the phenomenon whereby a moving object (traveling at a very high speed) will appear shorter to a stationary observer than it would have if at rest.

The relationship between the non-relativistic and relativistic length is given as follows:
<center> $l = \frac{l_0}{γ}$,

<p align="justify"> where $l_0$ is the length measured in the moving frame. Since the factor gamma is always greater than 1, this formula shows that length contracts (gets shorter) in the direction of motion. For instance, if a meter stick was thrown past you traveling at relativistic speed parallel to its length, the stick would look shorter than one meter long to you.

<p align = "justify"> Here's a similar question. If a meter stick was thrown past you traveling at relativistic speed perpendicular to its length, would the stick look shorter or longer than one meter to you? It would actually still look exactly one meter long to you because the direction of motion is not in the direction of length.

<p align = "justify"> Let's code another example and plot it. In the below code snippet, we define the proper length and use it to find the contracted length using the formula given above. Then, we plot contracted length as a function of velocity. You can test the code snippet out for different values of the proper length!


````python
def length_contraction(v, l_proper, c):
  gamma = 1/np.sqrt(1-(v**2/c**2))
  return l_proper/gamma

v_values = np.linspace(1, 0.99*c, 100)
l_proper = 1
l_contracted = length_contraction(v_values, l_proper, c)

plt.plot(v_values /c, l_contracted, label = "Length Contraction")
plt.xlabel('Velocity as a Fraction of Speed of Light')
plt.ylabel('Contracted Length (m)')
plt.legend()
plt.show()
````

# Special Relativity: Relativistic Momentum and Energy
Classical momentum and energy also don't hold at high speeds. If t' represents the time measured in the rest frame of an object, the object's momentum can be expressed as

<p align = "center"> $p = m_0 \frac{dx}{dt'}$.

In the frame of reference S with time t, this corresponds to

<p align = "center"> $p = m_0 \frac{dt}{dt'} \frac{dx}{dt} = m_0 \frac{dt}{dt'} v$.

Using the relationship we found for time dilation, $Δt = γ*Δt_0$, we can express $\frac{dt}{dt'}$ as γ. So, the formula for relativistic momentum becomes

<p align = "center"> $p = γm_0v$.

(Also, since Newton's Second Law states that $ F = \frac{d p}{dt}$, $F = \frac{d}{dt} γm_0v$.)

Work is the change in kinetic energy, and relativistic kinetic energy can be expressed as
<p align = "center"> $KE = γm_0c^2 - m_0c^2$.

<p align = "justify"> From this formula, we can see that as the speed of an object approaches the speed of light, gamma approaches infinity, meaning that an infinite amount of work would be needed to accelerate an object to the speed of light. Additionally, for speeds that are much lower than c, this relativistic form of kinetic energy reduces to $KE = \frac{1}{2} m_0 v^2$, which is the classical kinetic energy formula.

We can define the total energy of an object E as $E = γm_0c^2$. So,
<p align = "center"> $E = γm_0c^2 = KE - m_0c^2$.

Then, the total energy of an object is its kinetic energy plus the energy it has when at rest (i.e. its rest energy). (This is Einstein's famous equation $E = m_0c^2$, which is also known as 'Energy-Mass Equivalence'! It shows that even a small amount of mass can provide a huge amount of energy.)

The Energy-Momentum relation is defined as
<p align = "center"> $E^2 = p^2c^2 + m_0^2c^4 $.

One interesting thing here is that particles with no mass, such as photons, will still have momentum.

Let's code a snippet comparing classically-defined momentum and relativistic momentum for different values of v. When v is much less than c, the two equations for momentum should yield similar results; however, they will differ as v begins to approach c.

````python
def rel_momentum(v, classic_momentum, c):
  gamma = 1/np.sqrt(1-(v**2/c**2))
  return gamma*classic_momentum

v_values = np.linspace(1, 0.99*c, 100) #range of velocity values for graphing
m = 1 #mass of the object
classic_momentum = m*v_values


relative_momentum = rel_momentum(v_values, classic_momentum, 3e8)

plt.plot(v_values / 3e8, relative_momentum, color = 'blue', label="Relativistic Momentum")
plt.plot(v_values / 3e8, classic_momentum, color='orange', linestyle='--', label="Classical Momentum")
plt.xlabel("Velocity as a Fraction of c")
plt.ylabel("Momentum (m/s)")
plt.legend()
plt.show()
````

<p align = "justify"> Let's also code a snippet that will allow us to calculate the rest energy and total energy of an object with mass m and velocity v. You can input values for m and v, and the code will print out the mass, velocity, rest energy, and total energy of the object.

````python
def rest_energy(m, c):
  return m*c**2

def rel_momentum(v, m, c):
  gamma = 1/np.sqrt(1-(v**2/c**2))
  return gamma*m*v

def total_energy(v,m,c):
  return np.sqrt(rest_energy(m, c)**2 + (rel_momentum(v, m, c)**2)*c**4)

m = 10  #kg, input a value > 0
v = 2e6 #m/s

print("Mass (kg):", m, ", Velocity (m/s):", v)
print("Rest Energy (J):", rest_energy(m, c))
#print("Rest Energy (eV):", rest_energy(m, c)/1.6e-19) #this would be how to display the energy value in eV instead
print("Total Energy (J):", total_energy(v, m, c) / 3e8)
````

# Special Relativity: Relativistic Addition

Note that for simplicity, we will be dealing with velocities in 1D motion.

<p align = "justify"> In classical motion, velocities add like regular numbers do, with forward being the positive direction. For instance, imagine you are on the Polar Express and throw a snowball from the top of the train. The velocity of the train relative to the Earth is v. The velocity of the snowball relative to the train is u', and the velocity of the snowball relative to Earth is u. Then,
<p align = "center"> $ u = v + u'$.

<p align = "justify"> However, this kind of addition does not work for light, or objects moving at high velocities. For example, if instead of a snowball, you were holding a flashlight on top of the Polar Express, based on classical velocity addition the light would be traveling at total velocity $ u = v + c$. But we know that light will move away from the train at speed c.

For one-dimensional relativistic velocity addition, we use
<p align = "center"> $u = \frac{v + u'}{1 + \frac{vu'}{c^2}}$.

Let's write a code snippet showing the difference between classical velocity addition and relativistic velocity addition using a plot, given inputs of v and u'.

````python
def classical_addition(v_values, u_prime, c):
  return v_values + u_prime

def relativistic_addition(v_values, u_prime, c):
  return (v_values + u_prime)/(1 + (v_values*u_prime)/(c**2))

v_values = np.linspace(0,3e8,100) #m/s
u_prime = 2e8 #m/s

classical_result = classical_addition(v_values, u_prime, c)
relativistic_result = relativistic_addition(v_values, u_prime, c)

plt.plot(v_values / 3e8, classical_result / 3e8, color = 'darkgreen', linestyle = 'dashed', label="Classical Total Velocity")
plt.plot(v_values / 3e8, relativistic_result / 3e8, color='maroon', linestyle='dotted', label="Relativistic Total Velocity")
plt.xlabel("Velocity v as a Fraction of c")
plt.ylabel("Total Velocity as a Fraction of c")
plt.legend()
plt.show()
````

# Special Relativity: Doppler Shift
<p align = "justify"> The speed of light doesn't change when considering relativistic velocity, but frequency and wavelength do. This phenomenon is called a Doppler shift, and occurs in any wave where there is relative motion between the light source and the observer.

<p align = "justify"> You have probably already witnessed an effect of the Doppler shift, in sound waves. Have you ever heard the sound of an ambulance or police siren while driving? As the ambulance drives past you, the pitch of the sound of the siren drops noticeably. You are perceiving a change in the frequency of the sound waves coming from the siren.

Screenshot_20-5-2025_221520_www.bing.com.jpeg

<p align = "justify"> As you can see, the sound waves are perceived to have a lower frequency when traveling away from an observer and a higher frequency when approaching an observer. For light, the concept is analogous. The observed wavelength is longer (this is called a <i>redshift</i>) than that emitted by the light source when the source moves away from an observer, and shorter (this is called a <i>blueshift</i>) when the source moves towards the observer.
The Doppler equations for wavelength and frequency can be given by
<p align = "center"> $\lambda_o = \lambda_s \sqrt{\frac{1+ \frac{v}{c}}{1- \frac{v}{c}}}$

and
<p align = "center"> $f_o = f_s \sqrt{\frac{1- \frac{v}{c}}{1+ \frac{v}{c}}}$

where $\lambda_o$ and $f_o$ are the observed wavelength and frequency, and $\lambda_s$ and $f_s$ are the source wavelength and frequency. Also note that the signs (+ and -) are different for both equations.

Let's code a snippet analyzing the redshift of a galaxy!

Therefore, studying this galaxy and others like it could help scientists understand more about how the universe developed and about early galaxy formation. In terms of proper distance (the actual distance today), GN-z11 is estimated to be about 32 billion light-years away due to the ongoing expansion of the universe.

Note: z is called the "redshift parameter"; $z = \frac{λ_o - λ_s}{λ_s}$, where $λ_o$ is the wavelength measured from Earth and $λ_s$ is the wavelength originally emitted by the galaxy.

<p align = "justify"> (Another fun fact: you may have heard of the Andromeda Galaxy before because it's one of the galaxies closest to us... the Andromeda Galaxy is actually blueshifted because it is moving towards us!)

# Spacetime Diagrams

<p align = "justify"> Spacetime diagrams are graphs showing the position of objects as a function of time, with the object's location over time represented by a line or curve. They can be used to visualize relativistic effects, such as why different observers perceive the same event occurring at different times and places. The most well-known type of these diagrams is known as a Minkowski diagram, developed in 1908.

<p align = "justify"> In a Minkowski diagram, position is on the x-axis and time is on the y-axis. By convention, the y-axis is scaled by a factor of c so that both axes have the same units. Spacetime trajectories are represented by "world lines". For example, say you were to plot a stationary object on a Minkowski diagram. The position does not change but the time does, so the world line will be a vertical line. If you were to plot an object moving at a constant velocity, the world line would have a slope (positive if the motion is in the positive direction, and negative if in the negative direction). The slope of the line would be the speed of the object that the observer witnesses.

One note on simultaneity: events that are simultaneous in one reference frame will not be simultaneous in another moving reference frame.

<p align = "justify"> Another note on spacetime: in relativity, everything is described by "spacetime"; basically, there are three spatial dimensions (x ,y, z) and one time dimension to create a 4D coordinate system. In Minkowski diagrams, this 4D system is condensed into a 2D system for easier visualization.

<p align = "justify"> Let's code a snippet to generate a Minkowski spacetime diagram. We can set up the light cone (the trajectory that light would take) as well as worldlines for a stationary observer and a moving observer. Note that for something moving at a speed less than c, that worldline will have a steeper slope than the worldline of light.

````python
#Setting up the plot framework
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xlim(-9, 9)
ax.set_ylim(-9, 9)
ax.set_xlabel("Space (x)")
ax.set_ylabel("Time (ct)")
ax.axhline(0, color="black", linewidth=1)
ax.axvline(0, color="black", linewidth=1)

#light cone
x = np.linspace(-9, 9, 100)
ax.plot(x, x, 'r--', label="Speed of Light (c)")
ax.plot(x, -x, 'r--')

#Plotting the worldlines
ax.plot([2, 2], [-9, 9], 'b', label="Stationary Observer")

v = 0.6
x_moving = np.linspace(-9,9,100)
ax.plot(x_moving, x_moving/v, 'g', label="Moving Observer (v = 0.6c)")

ax.legend()
ax.set_title("Minkowski Spacetime Diagram with Lorentz Transformation")

plt.show()
````

# (Brief) Introduction to General Relativity

<p align = "justify"> Remember how you were taught in high-school physics about the "force of gravity"? Well, that may have been a bit misleading. The main concept of general relativity is that gravity isn't a force, but actually the curvature of spacetime.

<p align = "justify"> To understand what the "curvature of spacetime" even means, imagine the universe as a really large sheet that is being held taut on either side by a person. Now imagine dropping a tennis ball onto the sheet. The sheet will curve around the ball. What if you dropped a bowling ball onto the sheet? The sheet will curve a lot! So all objects will curve spacetime, and more massive objects will curve spacetime more.

<p align = "justify"> While on the topic of spacetime, one of the main consequences of special relativity is that space and time are viewed as <i> not </i> inherently distinct from each other (this is why we need to use a 4D coordinate system instead of thinking about space and time as separate things). To describe something in spacetime, you need a coordinate system. However, the laws of physics should not depend on the coordinate system you use, because the laws of physics should be the same for everyone. This is called the <i> principle of general covariance </i>.

<p align = "justify"> Sources of gravity (things that curve spacetime) are described by the <i> Einstein field equations </i>. These describe how things like energy, momentum, pressure, and stress (describes how both energy and momentum are distributed as well as how they flow from place to place) cause the curvature of spacetime.

# Black Holes and Relativity
<p align = "justify"> Say astronauts A and B are in space, and B is falling towards a black hole while A watches in horror from a safe distance. As B moves closer and closer to the black hole, to A, time seems to slow dramatically. That is, B seems to be slowing down as they get closer to the black hole. And, at the "event horizon" itself ($r/r_s = 1$), to A, B would appear to be frozen in time at that point. However, for B, they don't notice anything unusual until feeling extreme forces very close to the black hole. This phenomenon occurs due to the strength of the black hole's gravitational field; the gravitational time dilation is a result of the curvature of spacetime near a black hole.


Let's code a snippet showing the effects of time dilation near a non-rotating black hole!

<p align = "justify"> The Schwarzchild metric is a solution of the Einstein field equations that describes the gravitational field around a spherical object, assuming that the object does not have a charge and is not rotating. The Schwartzchild radius, used in this snippet, is the radius of a spherical object such that, if all the mass of that object was to be compressed within that radius, the escape velocity would equal the speed of light. So, any object with a radius smaller than its Schwarzchild radius will be a black hole.

The Schwarzchild radius is given by $r_s = \frac{2GM}{c^2}$, where G is the gravitational constant and M is the mass of the spherical object.

````python
G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
M1 = 2e30  # Mass of the black hole (kg) (close to sun's mass)

# Schwarzschild radius
r_s1 = (2 * G * M1) / c**2

# Generate distances from the black hole
r_values_1 = np.linspace(0, 15 * r_s1, 200)  # Distance from center outward

# Time dilation factor (from Schwarzschild metric)
time_dilation_factor_1 = 1/np.sqrt(1 - (r_s1 / r_values_1))

# Plot
plt.figure(figsize=(8,5))
plt.plot(r_values_1 / r_s1, time_dilation_factor_1, label="Relative Time Flow 1", color="darkgreen", linewidth=2)
plt.axvline(x=1, color="orange", linestyle="--", label="Event Horizon (r_s)")
plt.xlabel("Distance from Black Hole (in Schwarzschild radii)")
plt.ylabel("Time Flow (Relative to Distant Observer)")
plt.title("Time Dilation Near a Black Hole")
plt.legend()
plt.grid()

# Enhancements for readability
plt.xticks(np.arange(0, 16, step=2))  # Set readable ticks
plt.yticks(np.linspace(0, 5, 6))  # Normalize time dilation scale

plt.show()
````

<p align = "justify"> Another thing that happens to objects passing near a black hole, such as astronaut B, is spaghettification. Unfortunately, this occurrence is not nearly as tasty as it sounds. Spaghettification refers to the vertical stretching that occurs as an object approaches a black hole, because the gravitational pull on the near and far sides of the objects are different. If astronaut B were to fall feet-first into the black hole, their feet would feel a stronger pull than their head, stretching astronaut B into a long, thin shape (hence the name "spaghettification").

<p align = "justify"> (One note for clarification: the force that pulls apart an object is not the gravitational force itself, but the difference in the gravitational force between the ends of the object (ex. the feet and the head of astronaut B). This kind of force is called a <i> tidal force </i>. And yes, this is the same phenomenon that gives the Earth its tides.)

<p align = "justify"> Here's a question for you: we've seen that falling into a black hole with, say, the mass of the sun, would be deadly, but what if you were to fall into a supermassive black hole? For instance, the one at the center of our Milky Way galaxy? In fact, for the black hole at the center of the Milky Way, at the Schwarzchild radius (~ 12 million km), the gravity difference between head and feet would be about 0.02%; that is to say, barely noticeable*. Therefore, as far as spaghettification is concerned, larger black holes are actually less dangerous than smaller ones.

<p align = "justify"> Here's one last phenomenon: gravitational redshifting. This occurs when photons of emitted light are pulled towards a black hole. To escape, the light must "climb out" of the black hole's gravitational well. It releases energy in this process, and so the observed light has lower energy (and longer wavelength), resulting in a redder spectrum. Light passing the event horizon will go into the black hole and won't be able to escape. Observations of gravitational redshift around celestial bodies can give astronomers insights into properties of those bodies, such as mass and size.

This is why we don't perceive differences in gravity when we stand up. Luckily for you, on Earth the only kind of spaghettification you have to worry about is the one you're having for dinner.
